```python
import numpy as np
import pandas as pd
df = pd.read_csv('merged_stock_news_data.csv')
df
```

```python
df.dtypes
df['sentiment'].value_counts()
```

## Scaling columns except the headline columns

```python
from sklearn.preprocessing import MinMaxScaler
# Define which columns to scale
columns_to_scale = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'SMA_20', 'EMA_50', 'RSI', 
                    'Bollinger_High', 'Bollinger_Low', 'MACD']

# Initialize MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Apply scaling only to numerical columns
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
df
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['sentiment'] = label_encoder.fit_transform(df['sentiment'])
df
```

```python
from sentence_transformers import SentenceTransformer

# Load BERT-based sentence embedding model
bert_model = SentenceTransformer('bert-base-nli-mean-tokens')

# Convert headlines to embeddings
bert_embeddings = bert_model.encode(df['news_headline'].tolist())

# Convert to DataFrame
bert_df = pd.DataFrame(bert_embeddings, columns=[f"bert_vec_{i}" for i in range(bert_embeddings.shape[1])])

# Merge with dataset
df = pd.concat([df.drop(columns=['news_headline']), bert_df], axis=1)
```
